```python

df['Size'].value_counts().sort_index().plot(kind='bar')
plt.title('Count of Each Size (Sorted)')
plt.xlabel('Size')
plt.ylabel('Count')
plt.show()
```

```python
# Create box plots using seaborn
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Outlier Detection - Box Plots', fontsize=16)

# Age
sns.boxplot(y=df['Age'], ax=axes[0, 0])
axes[0, 0].set_title('Age')

# Purchase Amount
sns.boxplot(y=df['Purchase Amount (USD)'], ax=axes[0, 1])
axes[0, 1].set_title('Purchase Amount (USD)')

# Review Rating
sns.boxplot(y=df['Review Rating'], ax=axes[1, 0])
axes[1, 0].set_title('Review Rating')

# Previous Purchases
sns.boxplot(y=df['Previous Purchases'], ax=axes[1, 1])
axes[1, 1].set_title('Previous Purchases')

plt.tight_layout()
plt.show()
```
